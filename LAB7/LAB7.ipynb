```python
from PIL import Image, ImageDraw
import numpy as np
from google.colab import files
import io
from IPython.display import display

# Upload the image file
uploaded = files.upload()

if not uploaded:
    print("No file uploaded. Please upload the image file.")
else:
    image_name = list(uploaded.keys())[0]
    try:
        input_image = Image.open(io.BytesIO(uploaded[image_name]))

        if input_image.mode != 'RGB':
            input_image = input_image.convert('RGB')
            print("Đã chuyển đổi ảnh sang chế độ RGB.")

        # Convert to grayscale for edge detection (optional but recommended for Laplacian)
        gray_image = input_image.convert('L') # 'L' for grayscale
        input_pixels = gray_image.load()
        print("Đã chuyển ảnh sang ảnh xám để phát hiện cạnh.")

        # Define the Laplacian kernels [cite: image_3490bf.png]
        # Kernel a:
        laplacian_kernel_a = [[0, -1, 0],
                              [-1, 4, -1],
                              [0, -1, 0]]

        # Kernel b:
        laplacian_kernel_b = [[-1, -1, -1],
                              [-1, 8, -1],
                              [-1, -1, -1]]

        # List of kernels to apply
        kernels_to_apply = {
            "Kernel A (Laplacian)": laplacian_kernel_a,
            "Kernel B (Laplacian)": laplacian_kernel_b
        }

        # Process image for each kernel
        for kernel_name, current_kernel in kernels_to_apply.items():
            print(f"\n--- Áp dụng {kernel_name} ---")
            print(f"Kernel: \n{np.array(current_kernel)}")

            kernel = current_kernel
            offset = len(kernel) // 2
            output_edge_image = Image.new("L", input_image.size) # 'L' for grayscale output
            draw_edge = ImageDraw.Draw(output_edge_image)

            for x in range(offset, input_image.width - offset):
                for y in range(offset, input_image.height - offset):

                    pixel_sum = 0
                    for i in range(len(kernel)):
                        for j in range(len(kernel[0])):
                            neighbor_x = x - offset + i
                            neighbor_y = y - offset + j

                            if 0 <= neighbor_x < input_image.width and 0 <= neighbor_y < input_image.height:
                                pixel_value = input_pixels[neighbor_x, neighbor_y]
                                weight = kernel[i][j]
                                pixel_sum += pixel_value * weight

                    new_pixel_value = int(np.clip(abs(pixel_sum), 0, 255))
                    draw_edge.point((x, y), new_pixel_value)

            for x in range(input_image.width):
                for y in range(input_image.height):
                    if not (offset <= x < input_image.width - offset and
                            offset <= y < input_image.height - offset):
                        draw_edge.point((x, y), 0) # Set border to black

            output_image_path = f"{image_name.split('.')[0]}_{kernel_name.replace(' ', '_').replace('(', '').replace(')', '')}_edges.png"
            output_edge_image.save(output_image_path)
            print(f"Ảnh cạnh đã được lưu tại: {output_image_path}")
            print(f"Kết quả {kernel_name}:")
            display(output_edge_image)


        print("\nẢnh gốc:")
        display(input_image)


    except FileNotFoundError:
        print(f"Lỗi: Không tìm thấy tệp ảnh '{image_name}'. Vui lòng kiểm tra lại đường dẫn và tên file.")
    except Exception as e:
        print(f"Đã xảy ra lỗi: {e}")
```
